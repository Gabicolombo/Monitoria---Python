# Seleção dos dados

```python
produtos = df[["Produto", "Quantidade"]]
produtos
```

# Seleção com .loc

```python
display(df.loc[1:5])
```

```python
# buscar as linhas que corresponde uma condição específica, por exemplo produto 'Camiseta'

produto_camiseta = df.loc[df["Produto"] == 'Camiseta']
produto_camiseta
```

```python
produto_camiseta["subtracao"] = produto_camiseta["Valor Final"] - produto_camiseta["Valor Unitário"]
display(produto_camiseta)
```

```python
df["Comissão"] = df["Valor Final"] * 0.08

df
```

```python
df.loc[:, "Imposto"] = 0

df
```

# ordenando os dados do valor final - sort_values

```python
df_ordenado_cresc = df.sort_values(by="Valor Final")
df_ordenado_cresc
```

```python
df_ordenado_decresc = df.sort_values(by="Produto", ascending=False)
df_ordenado_decresc
```

```python
# groupby

faturamento = df[["Produto", "Valor Final"]].groupby("Produto").sum()
display(faturamento)
```

# Adicionando uma linha

```python
vendas_dez_df = pd.read_excel("Vendas - Dez (1).xlsx")

display(vendas_dez_df)
```

```python
display(df)
```

```python
vendas_df_concatecao = df.append(vendas_dez_df)

vendas_df_concatecao
```

```python
vendas_df_concatecao = vendas_df_concatecao.drop(0, axis=0)
vendas_df_concatecao
```

```python
# tirando os NaN
vendas_df_sem_na = vendas_df_concatecao.dropna(axis=1)
vendas_df_sem_na
```

```python
vendas_df_concatecao["Comissão"] = df["Comissão"].fillna(df["Comissão"].mean())
vendas_df_concatecao
```

```python
# value counts
transacoes_loja = vendas_df_concatecao["ID Loja"].value_counts()
display(transacoes_loja)
```

```python
vendas_df_gerentes = vendas_df_concatecao.merge(gerentes)

vendas_df_gerentes
```
